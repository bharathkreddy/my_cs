### Re-raise the error

````python
def divbyzero(a, b):
    try:
        return a/b
    except ZeroDivisionError as ex:
        print('log an error and continue...', str(ex))
        raise

divbyzero(1, 0)
````

````python
class CustomError(BaseException):
    """a custom exception"""
    pass

def divbyzero(a, b):
    try:
        return a/b
    except ZeroDivisionError as ex:
        print('log an error and continue...', str(ex))
        raise CustomError(*ex.args)

divbyzero(1, 0)
````

* All exceptions **inherit from** `BaseException`.
* Most user-facing exceptions derive from **`Exception`**, a subclass of `BaseException`.
* You can define **custom exceptions** by subclassing `Exception` or one of its descendants.

## ğŸ§© Core Concepts

### ğŸ”¹ BaseException vs Exception

* `BaseException` is the root of all exceptions â€” including system-level ones.
* **You should normally catch only `Exception`**, not `BaseException`, because:

  * `SystemExit`, `KeyboardInterrupt`, and `GeneratorExit` should not be intercepted unless explicitly intended.

---

### ğŸ”¹ Exception Categories

| Category            | Description                                                               |
| ------------------- | ------------------------------------------------------------------------- |
| **ArithmeticError** | Numeric calculation failures (`ZeroDivisionError`, `OverflowError`, etc.) |
| **LookupError**     | Container/key lookup failures (`IndexError`, `KeyError`)                  |
| **OSError**         | File system or I/O failures (permissions, missing files, etc.)            |
| **RuntimeError**    | Generic catch-all for invalid runtime states                              |
| **ValueError**      | Invalid argument values of the correct type                               |
| **TypeError**       | Operation applied to an inappropriate object type                         |
| **ImportError**     | Module import or loading failures                                         |
| **NameError**       | Name or variable not defined in scope                                     |
| **SyntaxError**     | Invalid Python syntax at parse time                                       |
| **ReferenceError**  | Attempt to access deleted weak references                                 |
| **MemoryError**     | Out of memory condition                                                   |
| **Warning**         | Non-fatal alerts used to signal deprecated or risky behavior              |

---

### ğŸ”¹ System-Level Exceptions (non-recoverable)

* `SystemExit`: Raised by `sys.exit()`
* `KeyboardInterrupt`: User pressed Ctrl+C
* `GeneratorExit`: Generator closed with `.close()`

---

### ğŸ”¹ Warnings (non-fatal subclasses of `Warning`)

Used for deprecation notices, risky operations, or environment issues.
They can be promoted to exceptions with `python -W error` or programmatically using the `warnings` module.

---

### ğŸ”¹ Built-in Exception Attributes

| Attribute                          | Description                                                    |
| ---------------------------------- | -------------------------------------------------------------- |
| `.args`                            | Tuple of arguments passed to the exception constructor         |
| `.with_traceback(tb)`              | Attaches a traceback object                                    |
| `.add_note(note)` *(Python 3.11+)* | Attaches explanatory text to the exception                     |
| `__cause__`                        | Reference to the original exception (via `raise ... from ...`) |
| `__context__`                      | Previous exception automatically chained                       |
| `__traceback__`                    | The traceback object for debugging                             |

---

### ğŸ”¹ Exception Chaining

Python automatically links exceptions that occur during exception handling:

* `__context__`: Automatically set when an exception happens inside another exception block.
* `__cause__`: Manually set when you raise a new exception using `raise NewError from OriginalError`.

---

### ğŸ”¹ Best Practices

* Always catch the **most specific** exception possible.
* Avoid `except Exception:` unless truly necessary.
* Use `finally` for cleanup (e.g., closing files or releasing locks).
* Prefer raising **custom exceptions** for domain-specific logic (e.g., `InvalidOrderError`, `DataParsingError`).
* Use the `warnings` module for non-fatal conditions instead of raising exceptions.

---

| Version   | Feature                                                                                                             |
| --------- | ------------------------------------------------------------------------------------------------------------------- |
| **3.11+** | `ExceptionGroup` and `BaseExceptionGroup` â€” handle multiple errors together (useful for asyncio or parallel tasks). |

---

### ğŸ§± Custom Exceptions

* Should inherit from `Exception` or a relevant subclass.
* Naming convention: end with **`Error`** (e.g., `ValidationError`, `TimeoutError`).
* Keep them lightweight, focused, and meaningful.
