# Anatomy of figures

1. `figure`: area where plot is rendered, container for everything
2. `axes`: inside figure, object for x and y axis
3. `ticks`: major and minor ticks
4. `major and minor tick labels`
5. `axis labels`
6. plot - the lines
7. legend
8. title

```python
x_pts_0 = np.linspace(-2 * math.pi, 2* math.pi, 100)
y_pts_0  = np.sin(x_pts_0)

ax[0].plot(x_pts_0, y_pts_0)

ax[0].set_title('Sin plot')
fig
```

```python
x_pts_1 = np.linspace(-2 * math.pi, 2* math.pi, 100)
y_pts_1  = np.tan(x_pts_1)

ax[1].plot(x_pts_1, y_pts_1)

ax[1].set_title('tan plot')
fig
```

# shared plot

```python
fig , ax = plt.subplots()

x_pts = np.linspace(-2*math.pi, 2*math.pi, 100)
ax.plot(x_pts, np.sin(x_pts), label = 'sin')
ax.plot(x_pts, np.cos(x_pts), label = 'cos')
ax.legend()
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_title('Trig-functions')
```

# shared axes

```python
fig, ax = plt.subplots(2, 1, sharex=True, figsize=(6,4))
x_pts = np.linspace(-2*math.pi, 2*math.pi, 100)

ax[0].plot(x_pts, np.sin(x_pts), label = 'sin')
ax[1].plot(x_pts, np.cos(x_pts), label = 'cos')
ax[0].set_title('sin')
ax[1].set_title('cos')

ax[1].set_xlabel('x')
ax[0].set_ylabel('sin(x)')
ax[1].set_ylabel('cos(x)')
```
