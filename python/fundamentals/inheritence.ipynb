````python
import numbers

class Rectangle:
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth
        self._area = None
        self._perimeter = None

    @property
    def length(self):
        return self._length

    @length.setter
    def length(self, length):
        if not isinstance(length, numbers.Number) or length < 0:
            raise ValueError(f'Length of {self.__class__.__name__} has to be a positive number')
        self._length = length
        self._area = None
        self._perimeter = None

    @property
    def breadth(self):
        return self._breadth

    @breadth.setter
    def breadth(self, breadth):
        if not isinstance(breadth, numbers.Number) or breadth < 0:
            raise ValueError(f'Breadth of {self.__class__.__name__} has to be a positive number')
        self._breadth = breadth
        self._area = None
        self._perimeter = None

    @property
    def area(self):
        if not self._area:
            self._area = self.length * self.breadth

        return self._area

    @property
    def perimeter(self):
        if not self._perimeter:
            self._perimeter = 2 * (self.length + self.breadth)

        return self._perimeter
        

class Square(Rectangle):
    def __init__(self, side):
        super().__init__(side, side)
    
````

````python
class IntegerField:
    def __set_name__(self, owner, name):
        self.private_name = f"_{name}"

    def __get__(self, instance, owner=None):
        if instance is None:
            return self
        return getattr(instance, self.private_name, 0)

    def __set__(self, instance, value):
        if not isinstance(value, int):
            raise TypeError(f"{self.private_name[1:]} must be an integer")
        setattr(instance, self.private_name, value)

class Point2D:
    x = IntegerField()
    y = IntegerField()

p1 = Point2D()
p1.x = 10
p1.y = 20

p2 = Point2D()
p2.x = 1
p2.y = 2
````
